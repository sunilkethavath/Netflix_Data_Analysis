# Remove error : time data " August 4, 2017" doesn't match format "%B %d, %Y", at position 218

```python
from datetime import datetime
def convert_to_datetime(date_str):
    if isinstance(date_str, str):
         try:
             date_string =date_str.strip()
             return datetime.strptime(date_str,"%B %d, %Y")
         except ValueError:
             return None
         
    else:
        return None
         

    # Import the datetime module


# Define the convert_to_datetime function


# Apply the convert_to_datetime function to the Release_Date column
data['Date_N'] = data['Release_Date'].apply(convert_to_datetime)

# Now, data['Date_N'] will contain datetime objects for valid dates and None for invalid ones

```
